```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df = pd.read_csv("../data/processed/final_hantavirus_dataset.csv")

print("Informasi Dataset Final:")
df.info()
display(df.head())
```

```python
plt.figure(figsize=(8, 6))
host_counts = df['host_label'].value_counts() 

sns.barplot(x=host_counts.index, y=host_counts.values, palette="viridis")
plt.title("Distribusi Kategori Inang (Host) Orthohantavirus", fontsize=14, fontweight='bold')
plt.xlabel("Kategori Inang", fontsize=12)
plt.ylabel("Jumlah Sekuens RNA", fontsize=12)
plt.xticks(rotation=45, ha='right') # Memutar label agar tidak saling tumpang tindih
plt.tight_layout()

plt.savefig("plot_host_distribution.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
seg_counts = df['segment_type'].value_counts()

sns.barplot(x=seg_counts.index, y=seg_counts.values, palette="cubehelix")
plt.title("Distribusi Tipe Segmen RNA (S, M, L)", fontsize=14, fontweight='bold')
plt.xlabel("Tipe Segmen", fontsize=12)
plt.ylabel("Jumlah Sekuens", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.savefig("plot_segment_distribution.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
geo_counts = df['geo_label_broad'].value_counts()

sns.barplot(x=geo_counts.index, y=geo_counts.values, palette="magma")
plt.title("Distribusi Regional / Geografis Asal Virus", fontsize=14, fontweight='bold')
plt.xlabel("Benua / Regional", fontsize=12)
plt.ylabel("Jumlah Sekuens", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.savefig("plot_geo_distribution.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='sequence_length', bins=50, kde=True, color="teal")
plt.title("Distribusi Panjang Sekuens RNA Orthohantavirus", fontsize=14, fontweight='bold')
plt.xlabel("Panjang Sekuens (Base Pairs)", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.tight_layout()

plt.savefig("plot_sequence_length.png", dpi=300)
plt.show()
```
